```
import pandas
import matplotlib.pyplot as plt

# pandas est la bibliothèque incontournable pour faire de l'analyse de données
# structure de données classique manipulée avec pandas : 'DataFrame' (données stockées selon 2 dimensions - lignes et colonnes)
# c'est donc parfaitement adapté aux données issues de pointages

# on importe le CSV dans un DataFrame
donnees = pandas.read_csv("https://raw.githubusercontent.com/codekodo/tracker-et-python/main/mvt_parabolique.csv", delimiter=";", keep_default_na=False)

# on a bien une struture DataFrame
print(type(donnees))

# affichage des données en DataFrame
print(donnees)

# conversion DataFrame > liste colonne par colonne
t = donnees['t'].values.tolist()
x = donnees['x'].values.tolist()
y = donnees['y'].values.tolist()

# affichage des données en listes
print('\n t : ', t)
print('\n x : ', x)
print('\n y : ', y)

# graphique
plt.plot(x, y, 'o', markersize = 4)
plt.xlabel("x (en m)")
plt.ylabel("y (en m)")
plt.axis('equal')
plt.title("Chronophotographie du mouvement")
plt.show()
```
